```

````python
df4 = df3.dropna(subset=["Year"])
df4
````

````python
df4['Year'].unique()
````

````python
df4['New_Year'] = df4['Date'] # create a new column with the same values as the original "Year" column
````

````python
df4['New_Year'].unique()
````

````python
if 'New_Year' in df4.columns:
    df5 = df4.loc[~df4['New_Year'].apply(lambda x: isinstance(x, float))]
df5
````

````python
df5['New_Year'].unique()
````

````python
df5
````

````python
df5["col_1"] = df5["New_Year"].str.extract("(\d{4})")
df5
````

````python
df4.loc[df4['New_Year'] > 1000, 'New_Year'] = None # set values higher than 1000 to None
df4.loc[df4['New_Year'] < 999, 'New_Year'] = df4.loc[df4['New_Year'] < 999, 'Date'] # replace values lower than 0 with values from the "Date" column
df4
````

````python
df4.loc[df4['New Year'] < 1000, 'Year'] = df4.loc[df4['Year'] < 1000, 'Date']
````

### 
With .loc method, we localizate the values that are =0 in Year Column, and if these conditions metets, the function replace this values with the ones that are in Date column.


````python
df4.loc[df4['Year'] == 0, 'Year'] = df4.loc[df4['Year'] == 0, 'Date']
df4
````
